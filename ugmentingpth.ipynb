```python
import numpy as np
```

```python
def pathfinder(a,residual):

    
    
    seen=[False]*len(a)
    queue=[0] #contains all the seen nodes
    seen[0]=True
    allpath={}
    while queue:
        last=queue.pop(0)
        for i,val in enumerate(residual[last]):
            if val>0:
                if seen[i]==False:
                    seen[i]=True
                    queue.append(i)
                    if a[last,i]>0:
                        allpath[i]=(last,'forward')
                    if a[i,last]>0:
                        allpath[i]=(last,'backward')
                    if i==len(a)-1:
                        key=list(allpath.keys())
                        back=key[-1]
                        path=[back]
                        while back!=0:
            
                            back=allpath[back][0]
                            path.append(back)
                        
                        return True, path,allpath
                    
    
    return False ,[], allpath
```

```python
def residualflowchange(a,x,residual,path,allpath):
    epsilon=max([max(a[i]) for i in range(len(a))])
    for i in range(len(path)-1):
        epsilon=min(epsilon,residual[path[i+1],path[i]])
    if epsilon>0:
        for i in range(len(path)-1):
            if allpath[path[i]][1]=='forward':
                x[path[i+1],path[i]]= x[path[i+1],path[i]]+epsilon
                x[path[i],path[i+1]]= -x[path[i+1],path[i]]
            
            if allpath[path[i]][1]=='backward':
                x[path[i+1],path[i]]= -(x[path[i],path[i+1]]-epsilon)
                x[path[i],path[i+1]]= -x[path[i+1],path[i]]
            
            residual[path[i+1],path[i]]= a[path[i+1],path[i]]-x[path[i+1],path[i]]
            residual[path[i],path[i+1]]= a[path[i],path[i+1]]-x[path[i],path[i+1]]
        print(path,epsilon)
        print('residual',residual)
        print('x',x)
    return x,residual
    
```

```python
def Optimizer(a):
    x=np.zeros((len(a),len(a)))
    residual=np.zeros((len(a),len(a)))

    result=pathfinder(a,residual)
    
    if result[0]==True:
        while result[0]:
            x=residualflowchange(a,x,residual,result[1],result[2])[0]
            residual=residualflowchange(a,x,residual,result[1],result[2])[1]
            result=pathfinder(a,residual)
        
    return x
        
```
